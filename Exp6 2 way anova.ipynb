```python
# Create a dataframe 
df = pd.DataFrame({'Watering': np.repeat(['daily', 'weekly'], 20),
                   'Sunlight': np.tile(np.repeat(['None','low', 'med', 'high'], 5), 2),
                   'height': [4.8, 5, 6.4, 6.3,4.4, 4.9, 5.8, 6,
4.4, 5.2, 6.2, 6.4,4.2, 5.3, 6.2, 4.9,

3.2, 5.6, 4.7, 5.6,3.8, 5.7, 6.3, 4.6,

3.9, 4.3, 5.5, 4.8,3.7, 5.4, 6.5, 5.6,

4.4, 4.8, 5.8, 5.8,3.9, 4.8, 5.5, 5.5 ]})
```

```python
df[:20]
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
#perform two-way ANOVA
model = ols('height ~ C(Sunlight) + C(Watering) + C(Watering):C(Sunlight)', data=df).fit()
result=sm.stats.anova_lm(model, typ=2)
print(result)
```

The watering p-value is equal to 0.435195
The Watering p-value is equal to 0.383331
The Fertilizer * Watering: p-value is equal to 0.1668238
The p-values for water and sun turn out to be less than 0.05 which implies that the means of both the factors possess a statistically significant effect on plant height. The p-value for the interaction effect (0.1668238) is greater than 0.05 which depicts that there is no significant interaction effect between fertilizer frequency and watering frequency.
